```python
#Carregando os Dados - Isso está feito para você! ^^
#Atenção, é importante que o arquivo 'lista_emails.csv' esteja no mesmo diretório que esse notebook!

import csv

lista = []
with open('lista_emails.csv', newline='') as inputfile:
    for row in csv.reader(inputfile):
        lista.append(row[0])
```
