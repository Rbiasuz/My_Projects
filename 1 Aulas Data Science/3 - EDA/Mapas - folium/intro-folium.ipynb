```python
import pandas as pd
import folium
```

Nosso primeiro mapa

```python
brasil = folium.Map(
    location=[-16.1237611, -59.9219642],
    zoom_start=4
)

# brasil.save('mapas_htmls/brasil_clean.html')
brasil
```

## Marcando pontos no mapa

Lendo os dados

```python
empresas = pd.read_csv('empresas.xz')
print(empresas.shape)
empresas.head(3)
```

Colocando algumas empresas no mapa

```python
brasil_com_pontos = folium.Map(
    location=[-29.161790, -51.150139],
    zoom_start=5,
    tiles='Stamen Terrain'
)

empresa1 = empresas.iloc[1]
folium.Marker(
    location=[empresa1['latitude'], empresa1['longitude']],
).add_to(brasil_com_pontos)

empresa2 = empresas.iloc[10]
folium.Marker(
    location=[empresa2['latitude'], empresa2['longitude']],
).add_to(brasil_com_pontos)

# brasil_com_pontos.save('mapas_htmls/brasil_some_companies.html')
brasil_com_pontos
```

 Então vamos escolher apenas um estado:

```python
empresas_rs= empresas[empresas['state'] == 'RS']
empresas.shape, empresas_rs.shape
```

```python
sample = empresas_rs.sample(150)
```

```python
for _, empresa in sample.iterrows():
    folium.Marker(
        location=[empresa['latitude'], empresa['longitude']],
    ).add_to(brasil)

brasil
```

Alguma cidade no litoral pernambucano

Colorindo de acordo com o bairro:

```python
colors = {
 'CAXIAS DO SUL': 'red',
 'PORTO ALEGRE': 'green',
 'SAO LEOPOLDO': 'orange',
 'SANTA MARIA': 'pink',
 'BENTO GONCALVES': 'purple',
 'GRAMADO': 'black',
}

brasil = folium.Map(
    location=[-29.161790, -51.150139],
    zoom_start=7
)

for _, empresa in sample.iterrows():
    if empresa['city'] in colors.keys():
        folium.Marker(
            location=[empresa['latitude'], empresa['longitude']],
            icon=folium.Icon(color=colors[empresa['city']])
        ).add_to(brasil)

brasil
```

```python
tooltip = 'Click me!'
folium.Marker([-29.161790, -51.150139], popup='<b>Estamos Aqui!!!</b>', tooltip=tooltip).add_to(brasil)
brasil
```
