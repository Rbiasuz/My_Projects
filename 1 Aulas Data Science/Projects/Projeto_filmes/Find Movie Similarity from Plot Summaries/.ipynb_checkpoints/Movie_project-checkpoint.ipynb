```python
# Começamos pelo básico, importando bibliotecas e olhando para nossos dados

import numpy as np
import pandas as pd
import nltk
import matplotlib.pyplot as plt

movies = pd.read_csv('datasets/movies.csv')

movies.head()
```

### Juntando as sinopses da Wikipedia e do IMD

Veja o exemplo da nossa primeira linha:
## The Godfather
* Wikipedia: "On the day of his only daughter's wedding, Vito Corleone"
* IMDb: "In late summer 1945, guests are gathered for the wedding reception of Don Vito Corleone's daughter Connie"

Ambas mostram informações relevates do filme, vamos então juntar elas e criar nossa feature "sinopse"

```python
movies['sinopse'] = movies['wiki_plot'].astype(str) + "\n" + movies['imdb_plot'].astype(str)
```

## Tf-idf and document similarity


(https://janav.wordpress.com/2013/10/27/tf-idf-and-cosine-similarity/)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
     'This is the first document.',
     'This document is the second document.',
     'And this is the third one.',
     'Is this the first document?'
        ]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
```

```python
vectorizer.get_feature_names()
```

```python
X.shape
```

```python
# Criando um especifico, que leve em conta "Stop-Words"

tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=200000,
                                   min_df=0.2, stop_words='english')


# load nltk's English stopwords as variable called 'stopwords'
stopwords = nltk.corpus.stopwords.words('english')
stopwords

stopwords
```

```python
# Aplicando o 'vetorizador' para nossas sinopses

tfidf_matrix = tfidf_vectorizer.fit_transform([x for x in movies["sinopse"]]) 
tfidf_matrix.shape
```

```python
# Clusterização

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 4, init='k-means++', random_state=1)
kmeans.fit(tfidf_matrix);
```

```python
# Criando uma coluna do nosso dataset chamada cluster, com a classificação do nosso modelo
clusters = kmeans.labels_.tolist()
movies["cluster"] = clusters

# Verificando o número de filmes por cluster
movies['cluster'].value_counts() 
```

```python
movies[['title', 'cluster']].head()
```

## Similaridade de textos

(https://en.wikipedia.org/wiki/Cosine_similarity)

```python
from sklearn.metrics.pairwise import cosine_similarity

teste = (
    "Macbook Pro - Nvidia GPU",
    "Macbook GPU",
    "Esse texto não tem nada com nada",
    "Oi! Eu sou o Goku!",
    "texto teste"
)

vetorizador_teste = TfidfVectorizer()
matrix_vetores_teste = tfidf_vectorizer.fit_transform(teste)

cosine_similarity(matrix_vetores_teste)
```

```python
# Calculate the similarity distance
similarity_distance = 1 - cosine_similarity(tfidf_matrix)

import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(similarity_distance, cmap='inferno');
```

```python
# Criando uma hierarquia
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
mergings = linkage(similarity_distance, method='complete')
# Plot the dendrogram
dendrogram_ = dendrogram(mergings,
               labels=[x for x in movies["title"]],
               leaf_rotation=90,
               leaf_font_size=16,
)

# Adjust the plot
fig = plt.gcf()
_ = [lbl.set_color('r') for lbl in plt.gca().get_xmajorticklabels()]
fig.set_size_inches(108, 21)

# Show the plotted dendrogram
plt.show()
```
