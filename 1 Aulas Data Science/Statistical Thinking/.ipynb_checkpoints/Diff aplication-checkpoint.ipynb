```python
#Packs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Parte 1: Coleta dos dados

```python
df = pd.read_html(r'https://www.worldometers.info/world-population/ireland-population/')[1]

#Font1: https://www.worldometers.info/world-population/ireland-population/
#Font2: http://worldpopulationreview.com/world-cities/dublin-population/
#Font3: http://population.city/ireland/dublin/
```

```python
df.to_csv('irland.csv')
```

```python
df
```

```python
plt.figure(figsize=(10,4))
plt.plot(df.Year,df.Population, label='Total Population')
plt.plot(df.Year, df['Urban Population'], label='Urban Population')
plt.legend();
```

```python
#Criando uma coluna com o número de anos considerando 1955 como ano 0
df['Tempo'] = df['Year']-1955

#Colocando a população na escala de 'milhões de pessoas'
df['Population'] = df['Population']/1000000
```

```python
plt.figure(figsize=(10,4))
plt.scatter(df.Tempo,df.Population)
plt.title('Scatter plot do total de população (M) pelo número de anos passados');
```

```python
#Separando os dados
X = df[['Tempo']]
y = df[['Population']]
```

## Parte 2: Modelagem

```python
# Abordagem linear (Usando SKLEARN)
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Ajustando os Dados ao nosso Modelo
model.fit(X, y)

# Predizendo os dados para calcular o erro posteriormente
y_pred = model.predict(X)

# Pontuando o Modelo
from sklearn.metrics import r2_score, mean_squared_error

# Valor de R2 perto de 1 nos diz que é um bom modelo
print(f"R2 score: {r2_score(y_pred, y)}")


# MSE Score perto de 0 é um bom modelo
print(f"MSE score: {mean_squared_error(y_pred, y)}")
```

```python
#Coeficientes lineares

print('Coeficientes: \n', model.coef_)
print('Intercept: \n', model.intercept_)
```

```python
#Equação do modelo:

            #coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])  
        
m = model.coef_[0]
b = model.intercept_
print('Equação do modelo:')
print('     y = {0} * x + {1}'.format(m.round(5), b.round(5)))
```

```python
# plotamos os resultados
plt.scatter(X, y,  color='black')
plt.plot(X, y_pred, color='blue', linewidth=3);
```

```python
# Abordagem polinominal (USANDO NUMPY e printando com SYMPY)
#https://docs.scipy.org/doc/numpy/reference/generated/numpy.poly1d.html
x = np.array(X)
x = x.reshape(-1)
y = np.array(y)
y = y.reshape(-1)
```

```python
#Calculando para os graus de 1 até 7

l_grau = []
l_coef = []
l_r2 = []
l_MSE = []
l_models = []

for grau in range(1,8):
    p = np.poly1d(np.polyfit(x, y, grau))

    print('GRAU: \n \n \n', grau, '\n \n \n')
    
    #Coeficientes
    coef = p.coefficients
    print('Coeficientes: \n', coef)
    
    print('Equação do modelo: \n')
    print(' y = 0 ')
    for i, v in enumerate(coef[::-1]):
        print('     + {:.5f} * x^{} '.format(v, i))
        
    # Valor de R2 perto de 1 nos diz que é um bom modelo
    print(f"R2 score: {r2_score(p(x), y)}")

    # MSE Score perto de 0 é um bom modelo
    print(f"MSE score: {mean_squared_error(p(x), y)}")
    
    
    # plotamos os resultados
    plt.scatter(x, y,  color='black')
    plt.plot(x, p(x), color='blue', linewidth=3)
    plt.show()
    
    l_grau.append(grau)
    l_coef.append(coef)
    l_r2.append(r2_score(p(x), y))
    l_MSE.append(mean_squared_error(p(x), y))
    l_models.append(p)
```

```python
#Armazendando resultados em um DataFrame
avaliacao = pd.DataFrame({
    "graus": l_grau,
    "r2": l_r2,
    "MSE": l_MSE,
    "coef": l_coef,
    "models": l_models
})
print(avaliacao)

(1-avaliacao['r2']).plot()
avaliacao['MSE'].plot()
plt.legend();
```

### Removendo 40% dos dados para avaliar os modelos

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.40, random_state=50)
```

```python
#Calculando para os graus de 1 até 7 e testando em novos dados

a_grau = []
a_coef = []
a_r2 = []
a_MSE = []
a_models = []

for grau in range(1,8):
    p = np.poly1d(np.polyfit(X_train, y_train, grau))

    print('GRAU: \n \n \n', grau, '\n \n \n')
    
    #Coeficientes
    coef = p.coefficients
    print('Coeficientes: \n', coef)
    
    print('Equação do modelo: \n')
    print(' y = 0 ')
    for i, v in enumerate(coef[::-1]):
        print('     + {:.5f} * x^{} '.format(v, i))
        
    # Valor de R2 perto de 1 nos diz que é um bom modelo
    print(f"R2 score: {r2_score(p(X_test), y_test)}")

    # MSE Score perto de 0 é um bom modelo
    print(f"MSE score: {mean_squared_error(p(X_test), y_test)}")
    
    
    # plotamos os resultados
    plt.scatter(x, y,  color='black')
    plt.plot(x, p(x), color='blue', linewidth=3)
    plt.show()
    
    a_grau.append(grau)
    a_coef.append(coef)
    a_r2.append(r2_score(p(X_test), y_test))
    a_MSE.append(mean_squared_error(p(X_test), y_test))
    a_models.append(p)
```

```python
#Armazendando resultados em um DataFrame
teste = pd.DataFrame({
    "graus": a_grau,
    "r2": a_r2,
    "MSE": a_MSE,
    "coef": a_coef,
    "models": a_models
})
print(teste)

(1-teste['r2']).plot()
teste['MSE'].plot()
plt.legend();
```

## Parte 3: Diferenciação da função para encontrar a taxa de crescimento (velocidade)

```python
import sympy as sp
sp.init_printing(pretty_print=True)
```

```python
#Criando a expressão
y1 = l_models[3]
x=Symbol('x')
expr = y1(x)
expr
```

```python
#Expandindo a expressão
sp.expand(expr)
```

```python
#Derivada da expressão
diff = expr.diff()
diff
```

```python
#Expandindo a derivada
diff.expand()
```

### Criando um intervalo de tempo de 1955 (inicio dos dados) até 2020 (final) para usar o modelo e extimar os seguintes parâmetros:

* População
* Taxa de Crescimento


```python
X = [i for i in range(2021-1955)]
X
```

```python
pop = []
taxa = []
for i in X:
    pop.append(expr.subs(x, i))
    taxa.append(diff.subs(x,i)*1000000) #Reconvertendo para Milhões de pessoas
```

```python
#Plotando os dois parâmetros
plt.figure(figsize=(10,4))
plt.subplot(121)
plt.plot(pop)
plt.title('População')
plt.subplot(122)
plt.plot(taxa)
plt.title('Taxa de Crescimento');
plt.tight_layout()
```

```python
# retornando os anos
X = [i+1955 for i in X]
```

```python
# Criando um DataFrame dos resultados
resultado = pd.DataFrame({
    "Ano": X,
    "População": pop,
    "Taxa de Crescimento": taxa
})
resultado.head()
```

```python
#Calculando a diferença populacional e de taxa de crescimento por ano
resultado['Diff Pop'] = resultado["População"].diff(1)
resultado['Diff Taxa'] = resultado["Taxa de Crescimento"].diff(1)
```

```python
#Plotando os dois novos parâmetros
plt.figure(figsize=(10,4))
plt.subplot(121)
plt.plot(resultado['Diff Pop'])
plt.title('Diff Pop')
plt.subplot(122)
plt.plot(resultado['Diff Taxa'])
plt.title('Diff Taxa');
plt.tight_layout()
```

```python
# arredondando os valores
resultado.dropna(inplace=True)
resultado.População = resultado.População.astype('float', inplace=True)
resultado['Taxa de Crescimento'] = resultado['Taxa de Crescimento'].astype('float', inplace=True)
resultado['Diff Pop'] = resultado['Diff Pop'].astype('float', inplace=True)
resultado['Diff Taxa'] = resultado['Diff Taxa'].astype('float', inplace=True)
```
