# Problema 2

```python
# EDA
x = np.array(tempo2)
y = np.array(bacterias)
plt.scatter(x,y)
```

```python
# Modelagem
grau = 1
p = np.poly1d(np.polyfit(x, y, grau))
```

```python
# Coeficientes
coef = p.coefficients
print('Coeficientes: \n', coef.round(3))
```

```python
# Plotamos os resultados
plt.scatter(x, y,  color='black')
plt.plot(x, p(x), color='blue', linewidth=3)
plt.show()
```

```python
# Avaliação do modelo
print('MSE = ', round(mean_squared_error(p(x),y),3))
print('MAE = ', round(mean_absolute_error(p(x),y),3))
```

```python
# Modelagem grau 2
grau = 2
p = np.poly1d(np.polyfit(x, y, grau))
# Coeficientes
coef = p.coefficients
print('Coeficientes: \n', coef.round(3))
# Plotamos os resultados
plt.scatter(x, y,  color='black')
plt.scatter(x, p(x), color='blue', linewidth=3)
plt.show()
# Avaliação do modelo
print('MSE = ', round(mean_squared_error(p(x),y),3))
print('MAE = ', round(mean_absolute_error(p(x),y),3))
```

```python
# Modelagem grau 3
grau = 3
p = np.poly1d(np.polyfit(x, y, grau))
# Coeficientes
coef = p.coefficients
print('Coeficientes: \n', coef.round(3))
# Plotamos os resultados
plt.scatter(x, y,  color='black')
plt.scatter(x, p(x), color='blue', linewidth=3)
plt.show()
# Avaliação do modelo
print('MSE = ', round(mean_squared_error(p(x),y),3))
print('MAE = ', round(mean_absolute_error(p(x),y),3))
```

```python
# Laço para avaliação:

for grau in range(2,16):
    p = np.poly1d(np.polyfit(x, y, grau))
    print('graus:', grau, 'MAE:', round(mean_absolute_error(p(x),y),3))
```

## Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x, y, 
    test_size=0.50, 
    random_state=42)
```

```python
# plotamos os resultados
plt.scatter(X_train, y_train, color='green', label='Train')
plt.scatter(X_test, y_test, color='red', label='Test')
plt.legend()
```

```python
#Calculando para os graus de 1 até 7 e testando em novos dados

lista_grau = []
lista_MAE = []
lista_models = []

for grau in range(2,16):
    p = np.poly1d(np.polyfit(X_train, y_train, grau))
   
    
    # plotamos os resultados
    plt.scatter(x, y,  color='black')
    plt.scatter(x, p(x), color='blue', linewidth=3)
    plt.show()
    
    lista_grau.append(grau)
    lista_MAE.append(round(mean_absolute_error(p(X_test),y_test),3))
    lista_models.append(p)
```

```python
plt.plot(lista_grau, lista_MAE, linestyle='-', linewidth=2)
plt.ylim(500,1000)
plt.scatter(lista_grau[np.argmin(lista_MAE)], min(lista_MAE), color='r', linewidth=4)
plt.plot()
```

```python
import pandas as pd
df = pd.DataFrame({
                    "graus": lista_grau,
                    "MAE": lista_MAE,
                    "Model": lista_models
                  })
df
```
