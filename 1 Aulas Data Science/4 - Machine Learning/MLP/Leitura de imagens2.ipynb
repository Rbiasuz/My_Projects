# Laço que leia diversos arquivos

```python
import os

folder = r'C:\Users\rodribia\Pictures\teste_imagens' # é so colocar aqui a pasta que quer utilizar
```

```python
# Lendo todos arquivos de uma pasta
for file in os.listdir(folder):
    print(file)
```

```python
# Manipulando strings
file.split('.')[0]
```

```python
# Juntando o diretorio com o arquivo
os.path.join(folder, file)
```

```python
lista_respostas = []
lista_arrays = []

for file in os.listdir(folder):
    lista_respostas.append(file.split('.')[0])
    
    arquivo = os.path.join(folder, file)
    im=cv2.imread(arquivo,0)
    
    resized = cv2.resize(im, (600,400))
    plt.figure(figsize=(8,5))
    plt.imshow(resized, cmap='gray');
    plt.axis('off')
    plt.show()
    
    array = resized.flatten()
    
    lista_arrays.append(array)
```

```python
lista_respostas
```

```python
len(lista_arrays)
```

```python
print('Número de pixels da última foto:', lista_arrays[4].shape)
print('Categoria da última foto:', lista_respostas[4])
print('Exemplo de pixels da última foto:', lista_arrays[4])
```

```python
df = pd.DataFrame(lista_arrays)
df['label'] = lista_respostas
df
```
