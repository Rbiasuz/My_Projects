## Groupby

#### O método groupby permite agrupar linhas de dados usando funções de agregação
* Média
* Soma
* Máx
* Min
* Etc...

Para mais informações, temos a documentação do método:
(https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html)

```python
#Primeiro, carregamos as bibliotecas:
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(np.concatenate((iris.data, np.array([iris.target]).T), axis=1), columns=iris.feature_names + ['target'])
df.head(7)
```

```python
agrupamento = df.groupby("target")
agrupamento
```

```python
agrupamento.mean()
```

```python
agrupamento.mean().T
```

```python
agrupamento.min()
agrupamento.max()
agrupamento.std()
```

```python
agrupamento.sum()
```

```python
agrupamento.describe()
```

```python
agrupamento.describe().T
```

```python
def my_function(df, n=10):
    return df.mean() * n
    
agrupamento.apply(my_function)
```

```python
agrupamento.apply(my_function, 100)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
agrupamento.boxplot(figsize=(20,15))
```

```python
agrupamento['sepal width (cm)'].plot()
plt.legend()
```

```python
agrupamento['sepal length (cm)'].plot()
plt.legend()
```

## Merging, Joining and Concatenating

#### Os três métodos são utilzados para "juntarmos" diferentes conjuntos de dados

* Por linhas
* Por Colunas
* Por Index
* Por uma chave de referência

### Concatenação

Junta um DataFrame no outro;

    Obs.: as dimensões devem corresponder ao longo do eixo que você está concatenando.

```python
#Vamos criar alguns dados para exemplificar:

df1 = pd.DataFrame({    'A': ['0', '1', '2', '3'],
                        'B': ['10', '11', '12', '13'],
                        'C': ['100', '111', '112', '113'],
                        'D': ['1000', '1001', '1002', '1003']})

df2 = pd.DataFrame({    'A': ['A', 'B', 'C', 'D'],
                        'B': ['AA', 'BB', 'CC', 'DD'],
                        'C': ['AAA', 'BBB', 'CCC', 'DDD'],
                        'D': ['AAAA', 'BBBB', 'CCCC', 'DDDD']})

df3 = pd.DataFrame({    'A': ['x0', 'x1', 'x2', 'x3'],
                        'B': ['y10', 'y11', 'y12', 'y13'],
                        'C': ['w100', 'w111', 'w112', 'w113'],
                        'D': ['q1000', 'q1001', 'q1002', 'q1003']})
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1,df2,df3], axis=0)
```

```python
pd.concat([df1,df2,df3], axis=0, ignore_index=True)
```

```python
pd.concat([df1,df2,df3], axis=1)
```

### Mescla

Mesclar permite que você una dois conjuntos de dados usando alguma coluna como referência

```python
df1 = pd.DataFrame({'coluna comum': ['Zelda', 'Mario', 'Sonic', 'Donkey Kong', 'Pokémon'],
                        'A': ['A0', 'A1', 'A2', 'A3', 'A4'],
                        'B': ['B0', 'B1', 'B2', 'B3', 'B4']})
    
df2 = pd.DataFrame({'coluna comum': ['Donkey Kong', 'Zelda', 'Sonic', 'Mario'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on=['coluna comum'])  #comum em ambas
```

```python
pd.merge(df1, df2, on=['coluna comum'], how='outer')  #Comum em apenas uma delas
```

### Join

Join tem um comportamento muito parecido com o método "merge", a diferença é que não  é preciso colocar colunas como "chave" para juntar os dataframes 

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['X', 'Y', 'Z']) 

df2 = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['X', 'Z', 'W'])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2) #Pega a primeira e junta os valores da segunda
```

```python
df2.join(df1)  #Pega a segunda e junta os valores da primeira
```

```python
df1.join(df2, how='outer')
```
