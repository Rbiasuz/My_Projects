```python
from tabula import read_pdf
import os
import pandas as pd
```

```python
tables = read_pdf('Descolamento AÃ§o + Killing + Passadeira 15d.is_tens.pdf', 
                  pages='all', encoding='ISO-8859-1', multiple_tables=True)
```

```python
cp1=[]
cp2=[]
cp3=[]
cp4=[]
cp5=[]
nome=[]
```

```python
dirname = 'descolamentos'
for file_name in os.listdir(dirname):
    if file_name.endswith('.pdf'):
        file_path = os.path.join(dirname, file_name)
        try:
            tables = read_pdf(file_path, pages='all', encoding='ISO-8859-1', multiple_tables=True)
            nome.append(file_name)
            cp1.append(tables[1].loc[5][6])
            cp2.append(tables[1].loc[6][6])
            cp3.append(tables[1].loc[7][6])
            cp4.append(tables[1].loc[8][6])
            cp5.append(tables[1].loc[9][6])
        except:
            print(file_name)
```

```python
d = {'nome':nome,'cp1':cp1,'cp2':cp2,'cp3':cp3,'cp4':cp4,'cp5':cp5}
df = pd.DataFrame(d)
df.head(8)
```

```python
df["nome"] = df["nome"].str.split('Descolamento ', expand = True)[1]
new = df["nome"].str.split("+", n=2, expand = True) 
df['Base']= new[0] 
df['Cola']= new[1]
df['Substrato']= new[2]   
df.drop(columns =["nome"], inplace = True) 
df["Substrato"] = df["Substrato"].str.split('72h', expand = True)
```

```python
df.to_csv('desc.csv', sep=';', encoding='ISO-8859-1')
```
